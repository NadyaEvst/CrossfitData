## Women athletes from Europe Regionals 2018 Data exploring

```python
import pandas as pd
#Create DataFrame from file containing men athletes from Europe Regionals 2018
athletesWomen = pd.read_csv("Data/athletesWomen.csv",sep=';',na_values = ['--'],index_col='CompetitorId');
```

```python
athletesWomen.head()
```

```python
%matplotlib inline
#Create DataFrame for identification of country names
labels = ['CountryCode','Country']
codeToCountryList=[('IS','Iceland'),
                  ('HU','Hungary'),
                  ('NO','Norway'),
                  ('SE','Sweden'),
                  ('PL','Poland'),
                  ('SK','Slovakia'),
                  ('GB','United Kingdom'),
                  ('LU','Luxembourg'),
                  ('FI','Finland'),
                  ('BE','Belgium'),
                  ('AT','Austria'),
                  ('DK','Denmark')]

countryCodeDF = pd.DataFrame.from_records(codeToCountryList,columns=labels)
#Create DataFrame with number of athletes by country-code
byCountry = athletesWomen.groupby('CountryCode')['CompetitorName'].count().reset_index(name ='Athletes on Regionals')
#Create DataFrame with number of athletes by country-code and country name
byCountry=pd.merge(countryCodeDF,byCountry,on='CountryCode')
byCountry
```

```python
toTheGames = athletesWomen.iloc[:5].groupby('CountryCode')['CompetitorName'].count().reset_index(name='Athletes going to The Games')
toTheGames = pd.merge(byCountry,toTheGames,on='CountryCode',how='left').fillna(0)
toTheGames
```

```python
ax = toTheGames.plot(x="Country",y="Athletes on Regionals",kind="bar",grid=True)
toTheGames.plot(x="Country",y="Athletes going to The Games",ax=ax,kind="bar",color="red",grid=True)
ax.set_title("Women athletes in Europe Regionals 2018 by countries")
```

```python
# Avarage body parameters for women athlete on Europe Regionals 2018
avarageAge = round(athletesWomen.Age.mean())
avarageHeight = round(athletesWomen.Height_cm.round().mean())
avarageWeight = round(athletesWomen.Weight_kg.round().mean())
avarageAthlete = {"Age": avarageAge,
                  "Height_cm":avarageHeight,
                 "Weight_kg":avarageWeight}
avarageAthlete
```

```python
# Avarage strength parameters for women athlete on Europe Regionals 2018
avarageBackSquat_kg = round(athletesWomen.BackSquat_kg.mean())
avarageCleanAndJerk_kg = round(athletesWomen.CleanAndJerk_kg.mean())
avarageSnatch_kg = round(athletesWomen.Snatch_kg.mean())
avarageDeadLift_kg = round(athletesWomen.DeadLift_kg.mean())

avarageStrength = {"BackSquat_kg": avarageBackSquat_kg,
                  "CleanAndJerk_kg":avarageCleanAndJerk_kg,
                 "Snatch_kg":avarageSnatch_kg,
                   "DeadLift_kg":avarageDeadLift_kg}
avarageStrength
```

```python
# Min strength parameters for women athlete on Europe Regionals 2018
minBackSquat_kg = round(athletesWomen.BackSquat_kg.min())
minCleanAndJerk_kg = round(athletesWomen.CleanAndJerk_kg.min())
minSnatch_kg = round(athletesWomen.Snatch_kg.min())
minDeadLift_kg = round(athletesWomen.DeadLift_kg.min())

minStrength = {"BackSquat_kg": minBackSquat_kg,
                  "CleanAndJerk_kg":minCleanAndJerk_kg,
                 "Snatch_kg":minSnatch_kg,
                   "DeadLift_kg":minDeadLift_kg}
minStrength
```

```python
# Count avarage stranght parameters by weight range:
# 1.Select columns with atheletes weight and strenght parameters
weightAndStranght = round(athletesWomen[['Weight_kg','BackSquat_kg', 'CleanAndJerk_kg', 'Snatch_kg',
       'DeadLift_kg']])
# 2.Assign every weight to weight range (range(75,110,5))
weightAndStranght['Weight_range'] = pd.cut(weightAndStranght.Weight_kg,range(55,105,5), include_lowest=True)
# 3. Now we don't need individual weights, let's drop that column
weightAndStranght=weightAndStranght.drop('Weight_kg',axis=1)
# 4. Groub by ranges and count avarage value for every strenght parameter
weightAndStranght = round(weightAndStranght.groupby('Weight_range').mean())
weightAndStranght
```
